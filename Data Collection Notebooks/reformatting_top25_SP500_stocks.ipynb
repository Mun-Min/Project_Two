```python
# Import dependencies
import pandas as pd
csv_path = "../Resources/merged_top25_sp500_stock_data.csv"
df = pd.read_csv(csv_path,index_col='date',
    parse_dates=True,
    infer_datetime_format=True)
df = df.drop(columns=['Unnamed: 0', 'date_utc', 'open','high','low', 'close','volume'])
df = df.pivot( columns='symbol',values='adjclose')
df
```

```python
df.to_csv('top25_sp500_stocks.csv')
```
