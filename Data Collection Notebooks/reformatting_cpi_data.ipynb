```python
#cpi_data.dtypes

cpi_data_list = pd.DataFrame()

for ind in cpi_data.index:
    for column in cpi_data.iloc[:,1:12]:
        cpi_data_list['date'] = (cpi_data[column].astype(str) + '-' + cpi_data['Year'].astype(str))
        cpi_data_list['value'] = cpi_data[column]
```
